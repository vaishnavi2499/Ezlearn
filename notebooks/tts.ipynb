```python
from TTS.api import TTS
import numpy as np

# Select the TTS model of your choice
model_path = "tts_models/en/ljspeech/tacotron2-DDC"  # Example model

# Load the TTS model
tts = TTS(model_path)

# Specify your text and speaker
text = "Hello world!"
speaker = "Claribel Dervla"  # Example speaker from your previous list

# Convert text to a format compatible with the TTS model
input_ids = tts.text_to_sequence(text)

# Create an attention mask (1 for real tokens, 0 for padding)
attention_mask = np.ones(len(input_ids), dtype=int)

# Convert text to speech with the specified speaker and attention mask
tts.tts_to_file(
    text=text,
    speaker=speaker,
    language="en",
    file_path="output.wav",
    attention_mask=attention_mask  # Include attention mask
)

print("Audio generated and saved as 'output.wav'")
```

```python
import torch
from TTS.api import TTS

# Load and initialize the TTS model
tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2")

# Use a predefined speaker ID or name if available (example speakers)
speaker_id = "Claribel Dervla"  # Replace this with an actual speaker ID from the model

# Generate speech as waveform data
wav = tts.tts(text="Hello world!", speaker=speaker_id, language="en")

# Save the output to a WAV file
output_path = "output.wav"
tts.tts_to_file(text="Hello world!", speaker=speaker_id, language="en", file_path=output_path)


# Define the model path for saving
model_path = r"C:\Users\VAISHNAVI\OneDrive\Desktop\dysproj\dys_pro\saved_models\coqui_tts_model.pth"

# Save the model state dictionary
torch.save(tts.state_dict(), model_path)
print(f"Model saved to {model_path}")

loaded_tts = TTS("tts_models/multilingual/multi-dataset/xtts_v2")  # Initialize with the same model structure
loaded_tts.load_state_dict(torch.load(model_path))  # Load the saved state
loaded_tts.eval()  # Set to evaluation mode
```
